````python
import pandas as pd
import numpy as np
import json
import requests
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Image
````

````python
df = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
````

````python
df = df[['Date', 'Time', 'Location', 'Operator', 'Aboard', 'Fatalities', 'Ground']]
df
````

````python
def findmilitary(string):
    if "Military" in string:
        return "Yes"
    else:
        return "No"
````

````python
df['Hour'] = df['Time'].apply(lambda x: str(x)[:2])
df['Month'] = df['Date'].apply(lambda x: str(x)[:2])
df['Decade'] = df['Date'].apply(lambda x: ((int(str(x)[-4:])//10)*10))
df['Military'] = df['Operator'].apply(lambda x: findmilitary(str(x)))
````

```Markdown
### As wars were more industrialized in the 60s 70s, I was wondering if crashes that is affliated with military purposes account for a big number of airplane crashes or not. I also wanted to see how the number of crashes is distributed through each decade as well as whether it has decreased or not.

### We can see from the graph that the number of crashes peaked in the 60s to 90s. The time period from the 60s - 90s is when the aviation field was developing and growing fiercely so the main causes for the majority of the crashes during the time was the lack of expertise in the field. Most major crashes result from airplane collisions both on the ground and on air (which is an error in ground aviation control), poor maintenance that leads to technical issues, ...

### Surprisingly, the number of crashes affliated with military only accounts for a small fraction of the total amount of crashes.
```

````python
labels = df.Decade.unique().tolist()
yesdata = []
nodata = []
for i in df.Decade.unique().tolist():
    yesdata.append((df[(df.Decade == i)&(df.Military == "Yes")].Military.size))
    
for i in df.Decade.unique().tolist():
    nodata.append(df[(df.Decade == i)].Military.size - df[(df.Decade == i)&(df.Military == "Yes")].Military.size)

fig, ax = plt.subplots(figsize=(8, 4))
width = 0.8 # bar width
xlocs = np.arange(len(yesdata))
ax.bar(xlocs, yesdata, width, color='#5885D2', label='Military')
ax.bar(xlocs, nodata, width, color='#A658D2', label='Non-Military', bottom=yesdata)
# --- labels, grids and title, then save
ax.set_yticks(range(0,900,50))
ax.set_xticks(ticks=np.array(range(len(yesdata))) + width/2)
ax.set_xticklabels(labels)
ax.set_xlim(-(1-width), xlocs[-1]+1)
ax.yaxis.grid(True)
ax.legend(loc='best')
ax.set_ylabel('Number of crashes')
fig.suptitle('Number of crashes in each decade, Military-specified')
````

```Markdown
### The last question I want to answer is the distribution of fatalities through decades. Similar to the amount of crashes, facilities peak in the 60s - 90s but reduced in the last few decades, suggesting an improvement in the aviation field. However, each decade we still face more than 10000 deaths, which is a scary number and suggests that we still need a lot of development in the field.

```

````python
x= df.Decade.unique().tolist()
y = []
for i in df.Decade.unique().tolist():
    y.append((df[(df.Decade == i)].Fatalities.sum()))
plots = plt.plot(x, y, label='')
plt.gcf().set_size_inches(8, 4)
plt.title('Fatalities through years')
plt.xlabel('Decade')
plt.ylabel('Fatalities')
plt.grid(True)
````
